#Import Library

```python
import pandas as pd
```

#Load Dataset

```python
url = 'https://raw.githubusercontent.com/Syukrondzeko/Fellowship/main/Seasons_Stats.csv'
data = pd.read_csv(url)
```

#Data Preprocessing

```python
#Remove Unnamed: 0 column
data = pd.read_csv(url, index_col=0)
data
```

```python
#Remove blank column name
del data["blanl"]
del data["blank2"]
data
```

```python
#Selecting player data in 2017
data_2017 = data[data['Year']==2017.0]
data_2017
```

```python
#Reset Index
data_2017 = data_2017.reset_index(drop=True)
data_2017
```

```python
#Drop duplicate data according to the player name
data_2017.drop_duplicates(subset ="Player",inplace = True)

#Reset index
data_2017 = data_2017.reset_index(drop=True)

data_2017
```

```python
#Identify null value
data_2017.isnull().sum()
```

```python
#Identify column data type
data_2017.info()
```

```python
#Replacing all null value with 0.0
data_2017.fillna(0.0, inplace = True)
data_2017.isnull().sum()
```

```python
#Convert data type
data_2017 = data_2017.astype({"Year": int, "Age": int, "FG": int, "FGA": int, "2P": int, 
                    "2PA": int, "FT": int, "ORB": int, "DRB": int, "TRB": int,
                    "AST": int, "STL": int, "BLK": int, "TOV": int, "PF": int,
                    "PTS": int})
data_2017.info()
```

#Question Number 1
###Who is the youngest and oldest in the NBA in 2017 for each team (Tm)?


```python
#Slicing data to get necessary column and value
data_young = data_2017[['Player','Tm','Age']].groupby(['Tm']).min().sort_values(by='Tm').reset_index()
data_old = data_2017[['Player','Tm','Age']].groupby(['Tm']).max().sort_values(by='Tm').reset_index()
```

```python
data_young
```

```python
data_old
```

#Question Number 2
###Which player has the most minutes played (MP) in each position (Pos)?


```python
#List of Most Minutes Played in each position
data_MP = data_2017.groupby('Pos')[['Player','MP']].max()

data_MP
```

#Question Number 3
###Which team has the highest average total rebound percentage (TRB%), assist percentage (AST%), steal percentage (STL%), and block percentage (BLK%)?


```python
#Set the average of each Team
data_highestaverage = data_2017[['Tm','TRB%','AST%','STL%','BLK%']].groupby(['Tm']).mean().reset_index()
data_highestaverage
```

```python
#The highest team average total rebound percentage (TRB%)
data_highestaverage[['Tm','TRB%']].sort_values(by='TRB%', ascending=False).head(1)
```

```python
#The gighest team average total rebound percentage (AST%)
data_highestaverage[['Tm','AST%']].sort_values(by='AST%', ascending=False).head(1)
```

```python
#The highest team average total rebound percentage (STL%)
data_highestaverage[['Tm','STL%']].sort_values(by='STL%', ascending=False).head(1)
```

```python
#The highest team average total rebound percentage (BLK%)
data_highestaverage[['Tm','BLK%']].sort_values(by='BLK%', ascending=False).head(1)
```

#Question Number 4
###Who is the best player in your opinion based on his record stats? note: you can refer to variables point (PTS), assists, rebounds, or anything else. A combination of several variables would be nice.

```python
#Set a new variable with overall value from attribute below
data_2017['Overall'] = data_2017.iloc[:,5:].mean(axis=1)
```

```python
#Top 10 best player based on overall value
data_2017[['Player','Overall']].sort_values('Overall', ascending=False).reset_index(drop=True).head(10)
```

#Question Number 5
###Which team has the best average stat record of their players? Note: you can refer to points, assists, rebounds, or anything else. A combination of several variables would be nice
